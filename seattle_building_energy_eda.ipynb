```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Admin\Documents\Energy_Consumption_Analysis\2015-building-energy-benchmarking.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.dropna(subset = ['ENERGYSTARScore'])
```

```python
df= df.fillna(df.median(numeric_only = True))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df.columns = (
    df.columns
    .str.lower()
    .str.replace('(', '', regex=False)
    .str.replace(')', '', regex=False)
    .str.replace('/', '_', regex=False)
    .str.replace('-', '_', regex=False)
)
```

```python
df.columns
```

```python
df.columns = [
    'building_type',
    'primary_property_type',
    'location',
    'neighborhood',
    'year_built',
    'number_of_buildings',
    'number_of_floors',
    'property_gfa_total',
    'property_gfa_buildings',
    'energy_star_score',
    'site_eui_kbtu_sf',
    'source_eui_kbtu_sf',
    'site_energy_use_kbtu',
    'steam_use_kbtu',
    'electricity_kbtu',
    'natural_gas_kbtu',
    'ghg_emissions_metric_tons_co2e',
    'ghg_emissions_intensity_kgco2e_ft2'
]
```

```python
df.to_csv("cleaned_energy_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['energy_star_score'], bins=20, kde=True)
plt.title('Distribution of Energy Star Scores')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['primary_property_type'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Property Types')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include=['int64', 'float64'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x=df['site_energy_use_kbtu'],
    y=df['ghg_emissions_metric_tons_co2e']
)

plt.title('Energy Use vs GHG Emissions')

plt.show()
```

```python
plt.figure(figsize=(12,6))

df.groupby('building_type')['energy_star_score'] \
  .mean() \
  .sort_values(ascending=False) \
  .plot(kind='bar')

plt.title('Average Energy Star Score by Building Type')
plt.ylabel('Average Score')
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

df.groupby('neighborhood')['site_energy_use_kbtu'] \
  .mean() \
  .sort_values(ascending=False) \
  .head(10) \
  .plot(kind='bar')

plt.title('Top Neighborhoods by Energy Usage')
plt.ylabel('Average Site Energy Use')

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['energy_star_score'])

plt.title('Energy Star Score Distribution')

plt.show()
```

```python
df['building_age'] = 2025 - df['year_built']
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x=df['building_age'],
    y=df['ghg_emissions_metric_tons_co2e']
)

plt.title('Building Age vs GHG Emissions')

plt.show()
```

```python
df.to_csv("cleaned_energy_data.csv", index=False)
```
